I wrote some tools to generate arbitrary models that I did not end up using, but here is a brief demo for how they work.

```python
from processor_connectivity_tools import connectivity_map

#initate with an undirected adjacency list
nairobi = connectivity_map([(0,1),(1,2),(1,3),(3,4),(4,5),(4,6)])
```

```python
#draw the topology of the processor using networkx
nairobi.draw_processor()
```

```python
#get all chains of connected nodes with length 3
nairobi.get_chains(chain_length = 3)
```

```python
#get all qubits 2 away from qubit 3
nairobi.connections_at_depth(3, 2)
```

```python
#generate a list of all paulis of weight <= max_weight on connected qubits
model_paulis = nairobi.proc_benchmark_max_weight(3)
print(model_paulis)
```
