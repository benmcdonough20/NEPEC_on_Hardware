# GST on Rigetti Backends with PyGSTi

```python
#PyGSTi tools
import pygsti
#pre-built gateset to use
from pygsti.modelpacks import smq1Q_XZ
from pygsti.data.dataset import DataSet
from pygsti.io.writers import write_dataset

#Rigetti tools
from pyquil import get_qc, Program
from pyquil.api import QCSClientConfiguration, local_forest_runtime
from pyquil.gates import RESET

configuration = QCSClientConfiguration.load()

#python helper libraries
from multiprocessing.pool import ThreadPool
import time
```

## Create experiment design

```python
target_model = smq1Q_XZ.target_model()      # a Model object
prep_fiducials = smq1Q_XZ.prep_fiducials()  # a list of Circuit objects
meas_fiducials = smq1Q_XZ.meas_fiducials()  # a list of Circuit objects
germs = smq1Q_XZ.germs()                    # a list of Circuit objects
maxLengths = [1,2,4,8,16,32]
exp_design = pygsti.protocols.StandardGSTDesign(target_model, prep_fiducials, meas_fiducials,
                                                germs, maxLengths)

exp_design.all_circuits_needing_data
circuits = list(exp_design.all_circuits_needing_data)
```

## Write empty protocol data file

```python
pygsti.io.write_empty_protocol_data('experiment_data/rigetti_first_run', exp_design, clobber_ok=True)
#Actual (`empty) dataset located at tutorial_files/test_gst_dir/data/dataset.txt
```

## Run circuits

```python
qpu = get_qc("Aspen-11", client_configuration=configuration, execution_timeout = 100000, compiler_timeout = 100000)
#qpu = get_qc("9q-square-qvm", noisy=True, execution_timeout = 100000, compiler_timeout = 100000)
```

```python
shots = 1000

quil_circuits = [Program(RESET(), circ.convert_to_quil()).wrap_in_numshots_loop(1000)
                 for circ in circuits]

def compile(prog):
    print("compiled")
    return qpu.compile(prog)

start_time = time.time()

with ThreadPool(12) as pool:
    executables = pool.map(compile, quil_circuits)

end_time = time.time()
print(end_time-start_time)
```

```python
results = []

def run(executable):
    result = qpu.run(executable).readout_data.get("ro")
    print("ran: ", result)
    zeros = len([i for i in result if i== [0]])
    return zeros

start_time = time.time()
    
with ThreadPool(12) as pool:
    results = pool.map(run, executables)
    
end_time = time.time()

print(end_time-start_time)
```

```python
time_data = [0]*len(results)
data = DataSet()

for (circ, result) in zip(circuits, results):
    data.add_count_dict(circ, {'0':result, '1':shots-result})
    
write_dataset("experiment_data/rigetti_first_run/data/dataset.txt", data)
```
